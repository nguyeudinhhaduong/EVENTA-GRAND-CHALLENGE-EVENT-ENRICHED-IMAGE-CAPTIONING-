```python
import numpy as np
import pandas as pd
import ast  # safer than eval

def to_np_array(embedding_col):
    cleaned = embedding_col.dropna().apply(
        lambda x: np.array(x if isinstance(x, list) else ast.literal_eval(x))
    )
    return np.vstack(cleaned.values)


# Convert embeddings from string to array
X_query = to_np_array(df_query['embedding'])       
X_base = to_np_array(df_Database_final['embedding'])  
print("X_query shape:", X_query.shape)
print("X_base shape:", X_base.shape)
```

```python
import faiss
# Build FAISS index
d = X_base.shape[1]  # dimensionality
index = faiss.IndexFlatL2(d)  # hoặc IndexFlatIP nếu cosine
index.add(X_base)

# Search top 10 nearest neighbors for each query
k = 10
distances, indices = index.search(X_query, k)

# Lấy id_article tương ứng từ df2
results = []
for i, idx_list in enumerate(indices):
    top_ids = df_Database_final.iloc[idx_list]['id_article'].values
    results.append({
        'query_image_path': df_query.iloc[i]['image_path'],
        'top_10_id_articles': top_ids.tolist()
    })

# Convert to DataFrame
result_df = pd.DataFrame(results)
```

## 1. Lấy 100 hình có điểm cosin simlarlity cao nhất (BM25_DF)

```python
import faiss
import numpy as np
import pandas as pd

# Normalize vector để dùng cosine similarity
X_base_norm = X_base / np.linalg.norm(X_base, axis=1, keepdims=True)
X_query_norm = X_query / np.linalg.norm(X_query, axis=1, keepdims=True)

# FAISS dùng Inner Product (IP) như là cosine similarity sau chuẩn hóa
d = X_base.shape[1]
index = faiss.IndexFlatIP(d)
index.add(X_base_norm)

k = 10
scores, indices = index.search(X_query_norm, k)  # điểm càng cao thì càng giống

results = []
for i, (idx_list, score_list) in enumerate(zip(indices, scores)):
    top_ids = df_Database_final.iloc[idx_list]['id_article'].values
    top_scores = score_list

    results.append({
        'query_image_path': df_query.iloc[i]['image_path'],
        'top_100_id_articles': top_ids.tolist(),           # list of article IDs
        'top_100_cosine_scores': top_scores.tolist()        # list of cosine scores
    })

bm25_df = pd.DataFrame(results)

```

## 4. BM25 cũ 

```python
from rank_bm25 import BM25Okapi
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import nltk
from nltk.corpus import stopwords

# 1. Chuẩn bị
nltk.download('punkt')
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

# 2. Hàm tiền xử lý văn bản
def tokenize_and_filter(text):
    tokens = nltk.word_tokenize(text.lower())
    return [word for word in tokens if word.isalnum() and word not in stop_words]

# 3. Hàm xử lý một dòng trong bm25_df
def process_bm25_row(row,df, df_chunks):
    top_ids = row['top_100_id_articles']
    cosine_scores = row['top_100_cosine_scores']

    # 3.1 Lọc từ df_chunks theo top_ids
    filtered_chunks = df_chunks[df_chunks['id_article'].isin(top_ids)].copy()

    # 3.2 Token hóa corpus và query
    tokenized_corpus = [tokenize_and_filter(text) for text in filtered_chunks['chunk_text']]
    tokenized_query = tokenize_and_filter(df['assistant_text'].iloc[0])


    # 3.3 Tính điểm BM25
    bm25 = BM25Okapi(tokenized_corpus)
    bm25_scores = bm25.get_scores(tokenized_query)
    filtered_chunks['bm25_score'] = bm25_scores

    # 3.4 Giữ mỗi id_article dòng có bm25_score cao nhất
    top_chunks = (
        filtered_chunks
        .sort_values('bm25_score', ascending=False)
        .drop_duplicates(subset='id_article', keep='first')
        .reset_index(drop=True)
    )

    # 3.5 Chuẩn hóa bm25_score
    if len(top_chunks) > 0:
        scaler = MinMaxScaler()
        top_chunks['bm25_score_scaled'] = scaler.fit_transform(top_chunks[['bm25_score']])
    else:
        top_chunks['bm25_score_scaled'] = []

    # 3.6 Tạo DataFrame cosine score
    cosine_df = pd.DataFrame({'id_article': top_ids, 'cosine_score': cosine_scores})
    cosine_df = (
        cosine_df
        .sort_values('cosine_score', ascending=False)
        .drop_duplicates(subset='id_article', keep='first')
    )

    # 3.7 Gộp điểm cosine vào
    top_chunks = top_chunks.merge(cosine_df, on='id_article', how='left')

    # 3.8 Tính điểm cuối cùng
    top_chunks['final_score'] = (
        top_chunks['cosine_score'] * 1.5+
        top_chunks['bm25_score_scaled'] * 0.2
    )

    # 3.9 Sắp xếp theo final_score
    top_chunks = top_chunks.sort_values('final_score', ascending=False).reset_index(drop=True)

    # 3.10 Trả top 10
    return top_chunks.head(10)

# 4. Ví dụ chạy cho dòng đầu tiên trong bm25_df
top_10_articles = process_bm25_row(bm25_df.iloc[0],caption_image , df_chunks)

# Xem kết quả
print(top_10_articles[['id_article', 'final_score', 'cosine_score', 'bm25_score_scaled']])
```

 ## 4. BM25 tập trung các tên riêng

```python
from rank_bm25 import BM25Okapi
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import nltk
from nltk.corpus import stopwords
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')  # Chỉ cần chạy 1 lần
stop_words = set(stopwords.words('english'))

# Tải tài nguyên cần thiết
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')  # giữ cái này
nltk.download('averaged_perceptron_tagger_eng')  # thêm dòng này để fix lỗi chính

# 2. Hàm token hóa với tăng trọng số tên riêng
def tokenize_and_filter_with_boost(text):
    tokens = nltk.word_tokenize(text)
    tagged = nltk.pos_tag(tokens)
    final_tokens = []
    for word, tag in tagged:
        word_lower = word.lower()
        if word_lower.isalnum() and word_lower not in stop_words:
            if tag in ('NNP', 'NNPS'):
                final_tokens.extend([word_lower]*3)  # tăng trọng số tên riêng
            else:
                final_tokens.append(word_lower)
    return final_tokens

# 3. Hàm xử lý một dòng trong bm25_df
def process_bm25_row(row, df, df_chunks):
    top_ids = row['top_100_id_articles']
    cosine_scores = row['top_100_cosine_scores']

    # 3.1 Lọc các chunk theo top_ids
    filtered_chunks = df_chunks[df_chunks['id_article'].isin(top_ids)].copy()

    # 3.2 Token hóa corpus và query
    tokenized_corpus = [tokenize_and_filter_with_boost(text) for text in filtered_chunks['chunk_text']]
    tokenized_query = tokenize_and_filter_with_boost(df['assistant_text'].iloc[0])

    # 3.3 Tính điểm BM25
    bm25 = BM25Okapi(tokenized_corpus)
    bm25_scores = bm25.get_scores(tokenized_query)
    filtered_chunks['bm25_score'] = bm25_scores

    # 3.4 Lấy mỗi id_article dòng có điểm cao nhất
    top_chunks = (
        filtered_chunks
        .sort_values('bm25_score', ascending=False)
        .drop_duplicates(subset='id_article', keep='first')
        .reset_index(drop=True)
    )

    # 3.5 Chuẩn hóa điểm BM25
    if len(top_chunks) > 0:
        scaler = MinMaxScaler()
        top_chunks['bm25_score_scaled'] = scaler.fit_transform(top_chunks[['bm25_score']])
    else:
        top_chunks['bm25_score_scaled'] = []

    # 3.6 Gộp điểm cosine
    cosine_df = pd.DataFrame({'id_article': top_ids, 'cosine_score': cosine_scores})
    cosine_df = (
        cosine_df
        .sort_values('cosine_score', ascending=False)
        .drop_duplicates(subset='id_article', keep='first')
    )

    top_chunks = top_chunks.merge(cosine_df, on='id_article', how='left')

    # 3.7 Tính điểm cuối cùng (final score)
    top_chunks['final_score'] = (
        top_chunks['cosine_score'] * 1.5 +
        top_chunks['bm25_score_scaled'] * 0.02 # tăng trọng số BM25
    )

    # 3.8 Trả top 10 kết quả
    return top_chunks.sort_values('final_score', ascending=False).reset_index(drop=True).head(10)

# 4. Ví dụ chạy cho dòng thứ 76 trong bm25_df
top_10_articles = process_bm25_row(bm25_df.iloc[76], caption_image, df_chunks)
print(top_10_articles[['id_article', 'final_score', 'cosine_score', 'bm25_score_scaled']])

```
